 # **<font color="DeepPink">Visualizaci√≥n del Pipeline de scikit-learn</font>**

<p align="justify">
üëÄ Un pipeline en scikit-learn es una forma conveniente de encadenar m√∫ltiples pasos de preprocesamiento y modelado en un √∫nico objeto. Un pipeline permite combinar varias etapas de procesamiento de datos, como la imputaci√≥n de valores faltantes, la codificaci√≥n de variables categ√≥ricas y el escalado de caracter√≠sticas, junto con un estimador final, como un modelo de aprendizaje autom√°tico, en una √∫nica secuencia de pasos.
<br><br>
Los pipelines son √∫tiles por varias razones:
<br>

1. **Facilitan el flujo de trabajo:** Con un pipeline, todas las transformaciones de datos y el ajuste del modelo se pueden realizar en un solo paso, lo que simplifica el c√≥digo y hace que el flujo de trabajo sea m√°s f√°cil de entender y de mantener.

2. **Evitan fugas de datos:** Al encapsular todas las transformaciones de datos y el modelo en un pipeline, se evita el riesgo de que se apliquen diferentes transformaciones a los datos de entrenamiento y de prueba, lo que puede provocar fugas de datos y resultados incorrectos.

3. **Permiten la optimizaci√≥n de hiperpar√°metros de forma conjunta:** Los pipelines tambi√©n facilitan la optimizaci√≥n de hiperpar√°metros de forma conjunta para todas las etapas del proceso, lo que puede llevar a mejores resultados de modelado.

<br>
<p align="justify">
‚úÖ Un pipeline en scikit-learn se define utilizando la clase <code>Pipeline</code> del m√≥dulo <code>pipeline</code>. Se especifica una lista de tuplas, donde cada tupla contiene el nombre de la etapa y el objeto correspondiente que implementa esa etapa (por ejemplo, un transformador de datos o un estimador).

‚ù§ https://scikit-learn.org/stable/

<p align="justify">
En este Colab desarrollaremos el Pipeline de <code>scikit-learn</code>. Primero cargamos un conjunto de datos.

```python
import numpy as np
import pandas as pd
```

```python
url="https://raw.githubusercontent.com/cristiandarioortegayubro/BDS/main/datasets/house_prices.csv"
```

```python
data = pd.read_csv(url, na_values="?")
```

```python
data.head()
```

```python
target_name = "SalePrice"
X = data.drop(columns=target_name)
y = data[target_name]
```

```python
X
```

```python
y
```

```python
y = (y > 200000).astype(int)
y
```

```python
data.head()
```

 # **<font color="DeepPink">Selecci√≥n de variables predictoras</font>**

<p align="justify">
üëÄ Seleccionamos algunas variables predictoras y generamos $3$ subsets.
</p>


```python
data.info()
```

```python
numeric_data = ["LotArea", "FullBath", "HalfBath"]
categorical_data = ["Neighborhood", "HouseStyle"]
```

```python
X = X[numeric_data + categorical_data]
X
```

 # **<font color="DeepPink">Creamos el Pipeline</font>**

 ## **<font color="DeepPink">Preprocesamiento de los datos</font>**

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy = 'median')),
    ('scaler', StandardScaler())])
```

```python
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
X.Neighborhood.value_counts()
```

 ## **<font color="DeepPink">Transformaciones de los datos</font>**

<p align="justify">
üëÄ El siguiente paso es aplicar las transformaciones usando <code>ColumnTransformer</code>

</p>


```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_data),
    ('cat', categorical_transformer, categorical_data)])
```

```python
preprocessor
```

 ## **<font color="DeepPink">Modelo de los datos</font>**

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())])
```

 # **<font color="DeepPink">Visualizamos el Pipeline</font>**

```python
model
```

 # **<font color="DeepPink">Evaluaci√≥n del modelo</font>**

```python
from sklearn.model_selection import cross_validate
```

```python
cv_results = cross_validate(model, X, y, cv = 5)
cv_results
```

```python
scores = cv_results['test_score']
```

```python
print("")
print("La media de Accuracy en la Validaci√≥n cruzada es: "
f"{scores.mean():.3f} +/- {scores.std():.3f}")
```
