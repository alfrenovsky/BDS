````python
departamentos.head()
````

<p>

````python
departamentos["CONTINENTE"] = "America"
````

````python
departamentos.head()
````

<p>
<font color="DeepPink">

> **Generar una columna que se llame PESOS cuyo valor sea DOLARES $\times$ $300$...**
</font>
</p>


````python
dolar = 300
departamentos["PESOS"] = departamentos["DOLARES"] * dolar
departamentos.head()
````

<p>
<font color="DeepPink">

> **Mostrar el resumen estadistico descriptivo del dataframe...**
</font>
</p>


````python
departamentos.describe().round(2)
````

````python
departamentos.describe().round(2).T
````

<p>
<font color="DeepPink">

> **Â¿Cual es la media de latitud y longitud?...**
</font>
</p>


````python
descriptivo = departamentos.describe()
````

````python
latitud = descriptivo.loc["mean","LATITUD"].round(2)
longitud = descriptivo.loc["mean","LONGITUD"].round(2)
````

````python
print(f"La media de las propiedades en latitud es {latitud} y la longitud {longitud}")
````
